```python
def clean_column_names(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
    return df
```

```python
clean_fb = clean_column_names(fb)
clean_ig = clean_column_names(ig)
```

The function clean_column_names wi;;remove all white spaces in the colunm names

```python
clean_ig["engagement_rate"] = ((clean_ig["shares"]+clean_ig['reel_shares']) / clean_ig["profile_reach"]) * 100
clean_ig.head()
```

The engagement rate is defined by the number of shares and reel_shares to the ratio of the profile it reached.

```python
clean_fb["date"] = pd.to_datetime(clean_fb["date"], errors="coerce")

# Extract 'Year-Month' for aggregation
clean_fb["year_month"] = clean_fb["date"].dt.to_period("M").astype(str)
monthly_date = clean_fb.groupby("year_month")
monthly_engagement = clean_fb.groupby("year_month")["page_post_engagements"].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_engagement, x="year_month", y="page_post_engagements", marker="o", color="b", linewidth=2)
plt.title("📈 Instagram Engagement Trends (Monthly)", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Avg. Engagement Rate (%)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
